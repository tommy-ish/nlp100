### 05. n-gram

```python
def n_gram(sequence, n):
    n_gram = []
    for i in range(0, len(sequence) - n + 1):
        n_gram.append(sequence[i:i + n])
    return n_gram
print(n_gram("I am an NLPer".split(" "), 2))
n_gram("I am an NLPer", 2)
```

### 06. 集合

```python
X = set(n_gram("paraparaparadise", 2))
Y = set(n_gram("paragraph", 2))
X | Y # 和集合
X & Y # 積集合
X - Y # 差集合
"se" in X
"se" in Y
```

### 08. 暗号文

```python
def cipher(string):
    s = ""
    for i in string:
        if i.islower():
            s += chr(219 - ord(i))
        else:
            s += i
    return s
print(cipher("I Am An Engineer."))
cipher("I An Am Emtrmvvi.")
```

### 09. Typoglycemia

```python
import random
words = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind.".replace(".", "").replace(":", "").split()
result = []
for word in words:
    if len(word) <= 4:
        result.append(word)
    else:
        characters = list(word[1:-1])
        random.shuffle(characters)
        result.append(word[0] + ''.join(characters) + word[-1])
result
```

```python
basic = re.sub(r"('{2,5})(.+)\1", r"\2", basic)
basic = re.sub(r"\[\[([^|]+?)\]\]", r"\1", basic)
print(re.sub(r"\[\[.+?\|(.+?)\]\]", r"\1", basic))
```

### 28. MediaWikiマークアップの除去

```python
basic = re.sub(r"\[\[.+?\|(.+?)\]\]", r"\1", basic)
basic = basic.replace("<br />", "")
basic = basic.replace("<ref>", "")
basic = basic.replace("</ref>", "")
basic = basic.replace('<ref name="imf-statistics-gdp">', "")
basic = basic.replace('<ref name="imf-statistics-gdp" />', "")
basic = basic.replace("<references/>", "")
basic = re.sub(r"\{\{lang\|.+?\|(.+?)\}\}", r"\1", basic)
basic = re.sub(r"\{\{center\|(.+?)\}\}", r"\1", basic)
basic = re.sub(r"\{\{Cite web\|url=http.+\|title=(.+?)\|.+\}\}", r"\1", basic)
basic = basic.replace("{{en icon}}", "")
basic = basic.replace("{{0}}", "0")
re.sub(r"\[.+\]", r"", basic)
print(basic)
```

### 30. 形態素解析結果の読み込み

```python
sentences = []
with open("/content/drive/MyDrive/neko.txt.mecab") as file:
    sentences = file.read().split("EOS")
import re
for i, sentence in enumerate(sentences):
    morphemes = []
    for morpheme in re.finditer(r"^(?P<surface>.*)\t(?P<pos>.+),(?P<pos1>.+),.+,.+,(?P<base>.+),.+,.+$", sentence, re.MULTILINE):
        morphemes.append(morpheme.groupdict())
    sentences[i] = morphemes
```

### 31. 動詞

```python
for sentence in sentences:
    for morpheme in sentence:
        if morpheme["pos"] == "動詞":
            print(morpheme["surface"])
```

### 32. 動詞の基本形

```python
for sentence in sentences:
    for morpheme in sentence:
        if morpheme["pos"] == "動詞":
            print(morpheme["base"])
```

### 33. 「AのB」

```python
for sentence in sentences:
    for i in range(1, len(sentence) - 1):
        if sentence[i - 1]["pos"] == "名詞" and sentence[i]["surface"] == "の" and sentence[i + 1]["pos"] == "名詞":
            print(sentence[i - 1]["surface"] + sentence[i]["surface"] + sentence[i + 1]["surface"])
```

### 34. 名詞の連接

```python
series = []
for sentence in sentences:
    nouns = []
    for morpheme in sentence:
        if morpheme["pos"] == "名詞":
            nouns.append(morpheme["surface"])
        else:
            if len(nouns) > 1:
                series.append("".join(nouns))
            nouns = []
    if len(nouns) > 1:
        series.append("".join(nouns))
series
```

### 35. 単語の出現頻度

```python
words = []
for sentence in sentences:
    for morpheme in sentence:
        words.append(morpheme["base"])
from collections import Counter
Counter(words).most_common()
```

### 37. 「猫」と共起頻度の高い上位10語

```python
cooccurrences = []
for sentence in sentences:
    words = [morpheme["surface"] for morpheme in sentence]
    if "猫" in words:
        cooccurrences.extend([word for word in words if word != "猫"])
plt.bar(*zip(*Counter(cooccurrences).most_common()[:10]))
plt.show()
```
